# Reading the content of URL and extract some information from it using "requests" library in python.

```python
import requests as req

url = 'https://www.lifewire.com/index-html-page-3466505'

response = req.get(url)

#200 is a standard response for successful HTTP requests and 404 tells that the requested resource could not be found.
print(response.status_code)

print("Returns the content of the response, in unicode")
print(response.text)

#Returns a JSON object of the result (if the result was written in JSON format, if not it raises an error)
#print(response.json())

print("Returns the header links")
print(response.links)

print("Returns the request object that requested this response")
print(response.request)
```
